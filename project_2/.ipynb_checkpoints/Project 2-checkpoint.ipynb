###  'NA' Value

```python
# select object dtypes

obj = df.select_dtypes(include='object')
```

```python
# Define threshold for 'NA' 

def threshold_na(x):
    remove_list = []
    for columns in obj:
        if ((obj[columns].str.contains('NA').sum())/len(obj)) > x: # number of 'NA' value > threshold
            remove_list.append(columns)
        
    return remove_list
```

```python
# Set number of 'NA' threshold at 80%

threshold_na(0.8)
```

```python
# Drop columns with high 'NA' according to threshold and less importance

df.drop(['alley','pool qc','misc feature','fence'],axis=1,inplace=True)
```

## Pre-processing

- One-hot encode categorical variables.
- Train/test split your data.
- Scale your data.
- Consider using automated feature selection.

```python
# Identify X,y

features = [col for col in df._get_numeric_data().columns if col != 'saleprice']
X = df[features]
y = df['saleprice']
```

```python
# Train test split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
# Scaling

ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

## Modeling

- **Establish your baseline score.**
- Fit linear regression. Look at your coefficients. Are any of them wildly overblown?
- Fit lasso/ridge/elastic net with default parameters.
- Go back and remove features that might be causing issues in your models.
- Tune hyperparameters.
- **Identify a production model.** (This does not have to be your best performing Kaggle model, but rather the model that best answers your problem statement.)
- Refine and interpret your production model.

```python
# Fit linear regression

lr = LinearRegression()
lr.fit(X_train_scaled,y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
lr.score(X_train_scaled,y_train)
```

```python
lr.score(X_test_scaled,y_test)
```

```python
# R2
lr_scores = cross_val_score(lr,X_train_scaled,y_train)
print(lr_scores.mean())

# MSE
lr_scores = -cross_val_score(lr,X_train_scaled,y_train,scoring='neg_mean_squared_error')
print(lr_scores.mean())
```

```python
# Ridge 
ridge = RidgeCV(alphas=np.linspace(.1,10,100))

# R2
ridge_scores = cross_val_score(ridge,X_train_scaled,y_train)
print(ridge_scores.mean())

# MSE
ridge_scores = -cross_val_score(ridge,X_train_scaled,y_train,scoring='neg_mean_squared_error')
print(ridge_scores.mean())
```

```python
# Lasso
lasso = LassoCV(n_alphas=200)

# R2
lasso_scores = cross_val_score(lasso,X_train_scaled,y_train)
print(lasso_scores.mean())

# MSE
lasso_scores_mse = -cross_val_score(lasso,X_train_scaled,y_train,scoring='neg_mean_squared_error')
print(lasso_scores_mse.mean())
```

```python
# Fit lasso to training data set 

lasso.fit(X_train_scaled,y_train)
```

```python
# train score
print(lasso.score(X_train_scaled,y_train))

# test score
print(lasso.score(X_test_scaled,y_test))

# lasso score R2
print(lasso_scores.mean())
```

```python
pd.Series(abs(lasso.coef_),index=features).sort_values(ascending= False).nlargest(25)
```
