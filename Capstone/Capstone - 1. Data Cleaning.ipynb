### Customer Reviews
Skytrax categorize reviews into "Trip verified", "not verified" and "verified review" tag in the customer_review columns which is not of importance to us and there are some special characters and logo appears before these words. Hence we will remove the tags along with special characters.

```python
# Remove special characters and "Trip Verified"
airline2['customer_review'] = airline2.loc[:,'customer_review'].map(lambda exp:exp.split('| ')[1] if "Trip Verified" in exp else exp)
```

```python
# Remove special characters and "not Verified"
airline2['customer_review'] = airline2.loc[:,'customer_review'].map(lambda exp:exp.split('|')[1] if 'not verified' in exp else exp)

# Remove special characters and "Verified review"
airline2['customer_review'] = airline2.loc[:,'customer_review'].map(lambda exp:exp.split('|')[1] if 'verified review' in exp else exp)
```

```python
# Sanity Check for review
airline2['customer_review'][1]
```

```python
# Change all words to lower caps
airline2[['airline', 'author', 'customer_review', 'traveller_type', 'cabin',
       'recommended']] = airline2[['airline', 'author', 'customer_review', 'traveller_type', 'cabin',
       'recommended']].apply(lambda x: x.str.lower())
```

```python
# Change recommended columns to binary (yes=1,No=0)
airline2['recommended'] = airline2['recommended'].map({'yes':1,'no':0})
```

```python
# Rename columns
airline2.columns = ['airline_name', 'author', 'date', 'content','type_traveller', 'cabin_flown', 'overall_rating',
       'seat_comfort_rating', 'cabin_staff_rating', 'food_beverages_rating','inflight_entertainment_rating', 'ground_service_rating',
       'value_money_rating', 'recommended']
```

## 6. Imputation

In order to maintain the integrity of the data as much as possible, we will be imputing based on the other available reviews of that passenger. 

The overall approach for imputation of this dataset is as per below:

|Level|Method|
|:---|:---|
|First Level|Impute using highly correlated columns
|Second Level|Impute using mean of all other available columns
|Third Level|Impute using mode based on airline's review (categorical)

![](./Image/imputation.png)

```python
# Heatmap of pearson correlations

heatmap_col = airline_final[['type_traveller','cabin_flown', 'overall_rating', 'seat_comfort_rating','cabin_staff_rating', 
              'food_beverages_rating','inflight_entertainment_rating', 'ground_service_rating','value_money_rating', 
              'recommended']]

fig,ax = plt.subplots(figsize=(14,12))

# Set the title.
ax.set_title('Airline Correlations', fontsize=25, y=1);

# Masking
mask = np.zeros_like(heatmap_col.corr())
mask[np.triu_indices_from(mask)] = True
ax = sns.heatmap(heatmap_col.corr(), mask=mask, vmax=0.8, square=True, annot = True, cmap='Blues')
```

### Recommended

Since there are some passengers that doesn't state their recommendations, hence we will use their overall rating scores to determine if they will recommend or not. Considering the scale of overall rating is from 1 to 10, we will assume if their overall rating is less than 5, they will not recommend and if their overall rating is 5 and above they will recommend. 

According to the calculation using the data, 95% of the reviews that has overall rating of more than 5 has recommended 'Yes' and overall rating of five and below has recommended 'No' hence it will be prudent to impute the recommended columns using overall ratings.

```python
airline_final.shape
```

```python
# Calculate accuracy of using overall rating to impute recommended
wrong_not_recommended = len(airline_final[(airline_final['overall_rating'] <= 5 ) & (airline_final['recommended'] == 1 )])
wrong_recommended = len(airline_final[(airline_final['overall_rating'] > 5 ) & (airline_final['recommended'] == 0 )])
1 - (wrong_recommended + wrong_not_recommended)/airline_final.shape[0]
```

```python
# Check if there are any reviews that has nulls in both  overall rating and recommended
airline_final[airline_final['recommended'].isnull() & airline_final['overall_rating'].isnull()]
```

```python
airline_final[(airline_final['recommended'].isnull()) & (airline_final['overall_rating'] > 5)].head()
```

```python
# Filter those overall rating 5 and above and impute with 1 in recommended
mask = (airline_final['recommended'].isnull()) & (airline_final['overall_rating'] > 5)
airline_final.loc[mask,'recommended'] = 1
```

```python
airline_final[(airline_final['recommended'].isnull()) & (airline_final['overall_rating'] <= 5)].head()
```

```python
# Filter those overall rating below 5 and impute with 1 in recommended
mask2 = (airline_final['recommended'].isnull()) & (airline_final['overall_rating'] <= 5)
airline_final.loc[mask2,'recommended'] = 0
```

```python
airline_final.isnull().sum()
```

### Value Money Rating

Since there is high correlations of 0.87 between overall ratings and value money ratings, hence we will impute using overall rating values.

The way to impute is using round up of `overall rating/2`. 

```python
# filter those with overall ratings
airline_final['value_money_rating'].fillna((airline_final['overall_rating']/2),inplace=True)

# Round up the ratings
airline_final['value_money_rating'] = airline_final['value_money_rating'].map(lambda x: math.ceil(x) if pd.notnull(x) else x)
```

```python
airline_final.isnull().sum()
```

### Ground Service Rating

Same for ground service rating, since there is high correlations of 0.85 with overall ratings, hence we will impute using overall rating values.

The way to impute is using round up of `overall rating/2`. 

```python
# filter those with overall ratings
airline_final['ground_service_rating'].fillna((airline_final['overall_rating']/2),inplace=True)

# Round up the ratings
airline_final['ground_service_rating'] = airline_final['ground_service_rating'].map(lambda x: math.ceil(x) if pd.notnull(x) else x)
```

```python
airline_final.isnull().sum()
```

### Cabin Staff Rating

There is high correlations of 0.80 with overall ratings, hence we will impute using overall rating values.

The way to impute is using round up of `overall rating/2`. 

```python
# filter those with overall ratings
airline_final['cabin_staff_rating'].fillna((airline_final['overall_rating']/2),inplace=True)

# Round up the ratings
airline_final['cabin_staff_rating'] = airline_final['cabin_staff_rating'].map(lambda x: math.ceil(x) if pd.notnull(x) else x)
```

```python
airline_final.isnull().sum()
```

### Cabin Flown

We will be imputing the null values by each individual airlines as it makes more sense. For categorical null values, we will impute using the most common values (mode). However, there are several airlines's cabin_flown that are completely null, hence we will filter those that can be imputed first.

```python
airline_final.head()
```

```python
# Group by airline and number of filled counts
cf_counts = airline_final.groupby('airline_name')['cabin_flown'].count()
```

```python
# airline's cabin_flown that is completely null
cf_counts[cf_counts == 0]
```

```python
# Filter and impute null values using mode
airline_final['cabin_flown'] = airline_final[airline_final.airline_name.isin(cf_counts[cf_counts > 1].index)].groupby('airline_name')['cabin_flown'].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# We will impute the remaining as economy class
airline_final['cabin_flown'].fillna('economy',inplace=True)
```

```python
# Check updated null
airline_final.isnull().sum()
```

### Type Traveller

For type of traveller, we will impute NA for the null values

```python
# We will impute the null as NA
airline_final['type_traveller'].fillna('NA',inplace=True)
```

```python
# Check updated null
airline_final.isnull().sum()
```

```python
airline_final.head()
```

### Impute the remaining null except overall rating using mean

The second level of imputing remaining null values with mean of each passenger's available reviews

```python
airline_final.iloc[:,7:13]
```

```python
airline_final[airline_final['inflight_entertainment_rating'].isnull()].head()
```

```python
# impute using the mean of the passenger's review
airline_final.iloc[:,7:13] = airline_final.iloc[:,7:13].T.fillna((airline_final.iloc[:,7:13].mean(axis=1))).T

# Round up the figures
airline_final.iloc[:,7:13] = airline_final.iloc[:,7:13].loc[:,:].applymap(lambda x: math.ceil(x) if pd.notnull(x) else x)
```

```python
# Check updated null
airline_final.isnull().sum()
```

```python
airline_final.loc[17:21,:]
```

### Overall Ratings
Lastly, we will impute the overall ratings with the mean of the other ratings.

```python
# Create Avg overall columns
airline_final['Avg_Overall'] = ((airline_final['seat_comfort_rating'] + airline_final['cabin_staff_rating'] + airline_final['food_beverages_rating'] + airline_final['inflight_entertainment_rating'] + airline_final['ground_service_rating'] + airline_final['value_money_rating'])/6)*2
airline_final['Avg_Overall'] = airline_final['Avg_Overall'].map(lambda x: math.ceil(x))
```

```python
# Impute the Overall using corresponding to Avg overall
airline_final['overall_rating'].fillna(airline_final['Avg_Overall'],inplace=True)
```

```python
# Check updated null
airline_final.isnull().sum()
```

```python
# Drop the avg overall
airline_final.drop('Avg_Overall',axis=1,inplace=True)
```

```python
# Change recommended type to int
airline_final['recommended'] = airline_final['recommended'].astype(int)
```

```python
# Change overall rating type to int
airline_final['overall_rating'] = airline_final['overall_rating'].astype(int)
```
