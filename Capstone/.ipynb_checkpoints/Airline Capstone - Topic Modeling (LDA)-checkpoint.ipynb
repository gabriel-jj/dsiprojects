```python
# Split the reviews in the content into good (recommend) or bad (not recommend):
good = [list(TextBlob(x.lower()).words) for x in tqdm(df[df['recommended']==1].content.values, desc='good')]
bad = [list(TextBlob(x.lower()).words) for x in tqdm(df[df['recommended']==0].content.values, desc='bad')]
```

### 1. Randomly assign words in documents to topics
- D represent document
- W represent words in documents
- Z represent collection of our K topics

```python
# Choose k topics:
K = 5
```

```python
# get the documents and all the words:
D = good+bad
W = [w for wordlist in D for w in wordlist]
print(len(D), len(W))
```

```python
# get the UNIQUE words:
W_unique = list(np.unique(W))
print(len(W_unique))
```

```python
# Make the topics. Each topic will have a dictionary of counts of occurrances
# of each word, which we will set to zero for now:
Z = {i:{w:0 for w in W_unique} for i in range(K)}
```

```python
Z.keys()
```

```python
# go through all the words in all the documents, and randomly assign each
# one to one of the topics:
D_z = [[[w, np.random.choice(range(K))] for w in d] for d in tqdm(D, desc='D')]
D_z[0]
```

```python
# update the topic dictionaries
for d in tqdm(D_z):
    for w,z in d:
        Z[z][w] += 1
```

### 2. Word distributions of topics

probability of a word being in a topic:
### $$ P(\text{word } w \;|\; \text{topic } z) $$

probability of *topic* occurance given a specific document:
### $$ P(\text{topic } z \;|\; \text{document } d) $$

```python
# p(z|d): proportion of words in document d assigned to topic z
def topic_given_doc(z_ind, d):
    return np.sum([w_z[1] == z_ind for w_z in d])/float(len(d))
```

```python
# p(w|z): proportion of words in topic z that are this word
def word_given_topic(w, z):
    total_words = np.sum(list(z.values()))
    if total_words == 0:
        return 0.
    else:
        return z[w]/float(total_words)
```

```python
topic_given_doc(3, D_z[0])
```

```python
word_given_topic('crew', Z[1])
```

### 3. Reassignment of topics

probability of a word in the specific document being generated from each of our topics
### $$ P(\text{topic } z \;|\; \text{document } d) \cdot P(\text{word } w \;|\; \text{topic } z) $$

```python
# calculate the the probability of a word in the first document from each topic
for z_ind in range(K):
    print('topic:', z_ind, 'P:', topic_given_doc(z_ind, D_z[0]) * word_given_topic('friendly', Z[z_ind]))
```

```python
def sample_topic(Z, document, word):
    probs = []
    for i in range(len(Z)):
        probs.append(topic_given_doc(i, document) * word_given_topic(word, Z[i]))
    if np.sum(probs) == 0:
        return -1
    return np.random.choice(range(len(Z)), p=np.array(probs)/np.sum(probs))
```

```python
sample_topic(Z, D_z[0],'friendly')
```

### 4. Getting words and topics

```python
def topic_dist(document, K=5):
    for z in range(K):
        print('topic:', z, 'P:', topic_given_doc(z, document))
```

```python
# Prob of topics in that document
for d, doc in enumerate(D_z[0:3]):
    print('document:', d)
    topic_dist(doc)
    print('-'*40)
```

```python
def topic_top_words(z, top=5):
    word_sum = np.sum(list(z.values()))
    word_counts = sorted(z.items(), key=lambda x: x[1], reverse=True)
    top_words = [[w, c/float(word_sum)] for w, c in word_counts[0:top]]
    for w, p in top_words:
        print(w, p)
```

```python
# Prob of word in that topic
for z_ind in range(K):
    print('topic:', z_ind)
    topic_top_words(Z[z_ind])
    print('-'*40)
```
